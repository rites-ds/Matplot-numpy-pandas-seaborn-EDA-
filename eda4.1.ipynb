## 📊 Integrated Business Data Generation

Creating a comprehensive business dataset that integrates:
- Customer demographics and behavior
- Sales performance across multiple dimensions
- Operational metrics and efficiency indicators
- Financial performance and profitability
- Market and competitive intelligence

```python
# Set random seed for reproducible executive reporting
np.random.seed(42)

# Generate comprehensive business intelligence dataset
current_date = datetime(2024, 8, 1)
start_date = datetime(2022, 1, 1)
date_range = pd.date_range(start=start_date, end=current_date, freq='D')

print("🏗️  GENERATING INTEGRATED BUSINESS DATASET")
print("=" * 60)

# Business units and their performance characteristics
business_units = {
    'E-Commerce': {'revenue_base': 1000000, 'growth_rate': 0.15, 'margin': 0.25},
    'Retail Stores': {'revenue_base': 800000, 'growth_rate': 0.05, 'margin': 0.20},
    'B2B Sales': {'revenue_base': 1200000, 'growth_rate': 0.10, 'margin': 0.30},
    'International': {'revenue_base': 600000, 'growth_rate': 0.20, 'margin': 0.18},
    'Subscription': {'revenue_base': 400000, 'growth_rate': 0.25, 'margin': 0.40}
}

# Generate monthly business performance data
monthly_periods = pd.date_range(start=start_date, end=current_date, freq='M')
business_data = []

for period in monthly_periods:
    # Calculate time-based factors
    months_from_start = (period - start_date).days / 30.44
    seasonal_factor = 1 + 0.2 * np.sin(2 * np.pi * period.month / 12)
    holiday_boost = 1.3 if period.month in [11, 12] else 1.0
    
    for unit, metrics in business_units.items():
        # Base performance with growth trend
        base_revenue = metrics['revenue_base'] * (1 + metrics['growth_rate']) ** (months_from_start / 12)
        
        # Apply seasonal and market factors
        actual_revenue = base_revenue * seasonal_factor * holiday_boost * np.random.uniform(0.8, 1.2)
        
        # Calculate related metrics
        gross_profit = actual_revenue * metrics['margin'] * np.random.uniform(0.9, 1.1)
        customers = int(actual_revenue / np.random.uniform(150, 350))  # Varying customer value
        orders = int(customers * np.random.uniform(1.2, 2.5))  # Orders per customer
        marketing_spend = actual_revenue * np.random.uniform(0.08, 0.15)
        operational_costs = actual_revenue * np.random.uniform(0.15, 0.25)
        
        business_data.append({
            'date': period,
            'business_unit': unit,
            'revenue': round(actual_revenue, 2),
            'gross_profit': round(gross_profit, 2),
            'marketing_spend': round(marketing_spend, 2),
            'operational_costs': round(operational_costs, 2),
            'customers': customers,
            'orders': orders,
            'customer_acquisition_cost': round(marketing_spend / max(customers * 0.2, 1), 2),
            'average_order_value': round(actual_revenue / max(orders, 1), 2)
        })

# Create comprehensive business DataFrame
business_df = pd.DataFrame(business_data)
business_df['date'] = pd.to_datetime(business_df['date'])
business_df['year'] = business_df['date'].dt.year
business_df['month'] = business_df['date'].dt.month
business_df['quarter'] = business_df['date'].dt.quarter

# Calculate additional KPIs
business_df['net_profit'] = business_df['gross_profit'] - business_df['marketing_spend'] - business_df['operational_costs']
business_df['profit_margin'] = business_df['net_profit'] / business_df['revenue'] * 100
business_df['marketing_roi'] = business_df['revenue'] / business_df['marketing_spend']
business_df['revenue_per_customer'] = business_df['revenue'] / business_df['customers']

print(f"📈 Generated business data: {len(business_df):,} records")
print(f"📅 Period coverage: {business_df['date'].min().strftime('%b %Y')} - {business_df['date'].max().strftime('%b %Y')}")
print(f"🏢 Business units: {', '.join(business_units.keys())}")
print(f"💰 Total revenue generated: ${business_df['revenue'].sum():,.0f}")

print("\n📊 Business Intelligence Dataset Overview:")
print(business_df.head())
```

## 📊 Executive KPI Dashboard

```python
# Calculate comprehensive executive KPIs
print("📊 EXECUTIVE KPI DASHBOARD")
print("=" * 70)

# Current period performance (last 3 months)
current_period = business_df[business_df['date'] >= (current_date - timedelta(days=90))]
previous_period = business_df[
    (business_df['date'] >= (current_date - timedelta(days=180))) & 
    (business_df['date'] < (current_date - timedelta(days=90)))
]

# Aggregate KPIs
current_metrics = current_period.groupby('business_unit').agg({
    'revenue': 'sum',
    'gross_profit': 'sum',
    'net_profit': 'sum',
    'customers': 'sum',
    'orders': 'sum',
    'marketing_spend': 'sum'
}).round(0)

previous_metrics = previous_period.groupby('business_unit').agg({
    'revenue': 'sum',
    'gross_profit': 'sum',
    'net_profit': 'sum',
    'customers': 'sum',
    'orders': 'sum',
    'marketing_spend': 'sum'
}).round(0)

# Calculate growth rates
growth_metrics = ((current_metrics - previous_metrics) / previous_metrics * 100).round(1)

# Overall company KPIs
total_current_revenue = current_metrics['revenue'].sum()
total_previous_revenue = previous_metrics['revenue'].sum()
revenue_growth = (total_current_revenue - total_previous_revenue) / total_previous_revenue * 100

total_current_profit = current_metrics['net_profit'].sum()
total_previous_profit = previous_metrics['net_profit'].sum()
profit_growth = (total_current_profit - total_previous_profit) / total_previous_profit * 100

print(f"🎯 COMPANY-WIDE PERFORMANCE (Last 90 Days vs Previous 90 Days):")
print(f"   💰 Total Revenue: ${total_current_revenue:,.0f} ({revenue_growth:+.1f}%)")
print(f"   📈 Net Profit: ${total_current_profit:,.0f} ({profit_growth:+.1f}%)")
print(f"   👥 Total Customers: {current_metrics['customers'].sum():,.0f}")
print(f"   🛒 Total Orders: {current_metrics['orders'].sum():,.0f}")
print(f"   💵 Overall Profit Margin: {(total_current_profit/total_current_revenue*100):.1f}%")

print(f"\n🏢 BUSINESS UNIT PERFORMANCE:")
print("=" * 70)

for unit in business_units.keys():
    if unit in current_metrics.index:
        current_rev = current_metrics.loc[unit, 'revenue']
        rev_growth = growth_metrics.loc[unit, 'revenue'] if unit in growth_metrics.index else 0
        current_profit = current_metrics.loc[unit, 'net_profit']
        profit_margin = (current_profit / current_rev * 100) if current_rev > 0 else 0
        
        print(f"\n📊 {unit}:")
        print(f"   Revenue: ${current_rev:,.0f} ({rev_growth:+.1f}%)")
        print(f"   Profit Margin: {profit_margin:.1f}%")
        print(f"   Customers: {current_metrics.loc[unit, 'customers']:,.0f}")

# Performance ratings
print(f"\n⭐ PERFORMANCE RATINGS:")
for unit in business_units.keys():
    if unit in growth_metrics.index:
        rev_growth = growth_metrics.loc[unit, 'revenue']
        if rev_growth >= 15:
            rating = "🟢 Excellent"
        elif rev_growth >= 5:
            rating = "🟡 Good"
        elif rev_growth >= 0:
            rating = "🟠 Fair"
        else:
            rating = "🔴 Needs Attention"
        print(f"   {unit}: {rating} ({rev_growth:+.1f}% growth)")
```

## 🎯 Strategic Business Intelligence & Recommendations

```python
# Comprehensive strategic analysis for executive decision-making
print("🎯 STRATEGIC BUSINESS INTELLIGENCE REPORT")
print("=" * 80)

# Performance benchmarking
current_year_data = business_df[business_df['year'] == 2024]
previous_year_data = business_df[business_df['year'] == 2023]

# Calculate year-over-year metrics
current_ytd_revenue = current_year_data['revenue'].sum()
previous_ytd_revenue = previous_year_data['revenue'].sum() * (8/12)  # Adjust for partial year
ytd_growth = (current_ytd_revenue - previous_ytd_revenue) / previous_ytd_revenue * 100

# Business unit performance analysis
unit_performance = business_df.groupby('business_unit').agg({
    'revenue': ['sum', 'mean', 'std'],
    'profit_margin': 'mean',
    'marketing_roi': 'mean',
    'customers': 'sum'
}).round(2)

unit_performance.columns = ['total_revenue', 'avg_monthly_revenue', 'revenue_volatility', 
                           'avg_profit_margin', 'avg_marketing_roi', 'total_customers']

# Calculate market share and growth potential
unit_performance['market_share'] = (unit_performance['total_revenue'] / 
                                   unit_performance['total_revenue'].sum() * 100)

unit_performance['growth_score'] = (
    (unit_performance['avg_marketing_roi'] - unit_performance['avg_marketing_roi'].mean()) / 
    unit_performance['avg_marketing_roi'].std() * 0.4 +
    (unit_performance['avg_profit_margin'] - unit_performance['avg_profit_margin'].mean()) / 
    unit_performance['avg_profit_margin'].std() * 0.6
)

print(f"📈 EXECUTIVE PERFORMANCE SUMMARY:")
print(f"   • YTD Revenue: ${current_ytd_revenue:,.0f} ({ytd_growth:+.1f}% vs last year)")
print(f"   • Total Profit: ${business_df['net_profit'].sum():,.0f}")
print(f"   • Customer Base: {business_df['customers'].sum():,} total customers")
print(f"   • Business Units: {len(business_units)} active units")

print(f"\n🏆 TOP PERFORMING BUSINESS UNITS:")
top_performers = unit_performance.nlargest(3, 'growth_score')
for i, (unit, metrics) in enumerate(top_performers.iterrows(), 1):
    print(f"   {i}. {unit}:")
    print(f"      • Revenue: ${metrics['total_revenue']:,.0f} ({metrics['market_share']:.1f}% share)")
    print(f"      • Profit Margin: {metrics['avg_profit_margin']:.1f}%")
    print(f"      • Marketing ROI: {metrics['avg_marketing_roi']:.1f}x")
    print(f"      • Growth Score: {metrics['growth_score']:.2f}")

print(f"\n⚠️  UNITS REQUIRING ATTENTION:")
attention_units = unit_performance.nsmallest(2, 'growth_score')
for unit, metrics in attention_units.iterrows():
    issues = []
    if metrics['avg_profit_margin'] < unit_performance['avg_profit_margin'].mean():
        issues.append("Low Profitability")
    if metrics['avg_marketing_roi'] < unit_performance['avg_marketing_roi'].mean():
        issues.append("Poor Marketing ROI")
    if metrics['revenue_volatility'] > unit_performance['revenue_volatility'].mean():
        issues.append("High Volatility")
    
    print(f"   • {unit}: {', '.join(issues) if issues else 'Performance Review Needed'}")

# Market opportunity analysis
print(f"\n🎪 MARKET OPPORTUNITIES:")
# Identify seasonality patterns
seasonal_performance = business_df.groupby('quarter')['revenue'].mean()
peak_quarter = seasonal_performance.idxmax()
low_quarter = seasonal_performance.idxmin()
seasonality_impact = (seasonal_performance.max() - seasonal_performance.min()) / seasonal_performance.mean() * 100

print(f"   • Seasonal Optimization: Q{peak_quarter} peak, Q{low_quarter} low ({seasonality_impact:.1f}% variance)")
print(f"   • Cross-Selling Potential: Leverage high-performing units")
print(f"   • Digital Transformation: Focus on E-Commerce growth")
print(f"   • International Expansion: Build on International unit success")

# Risk assessment
print(f"\n🛡️  RISK ASSESSMENT:")
overall_volatility = business_df.groupby('date')['revenue'].sum().std() / business_df.groupby('date')['revenue'].sum().mean() * 100
profit_stability = business_df.groupby('date')['profit_margin'].mean().std()

risk_factors = []
if overall_volatility > 20:
    risk_factors.append(f"High Revenue Volatility ({overall_volatility:.1f}%)")
if profit_stability > 5:
    risk_factors.append(f"Unstable Profit Margins (±{profit_stability:.1f}%)")
if ytd_growth < 5:
    risk_factors.append("Slow Growth Trajectory")

if risk_factors:
    for risk in risk_factors:
        print(f"   ⚠️  {risk}")
else:
    print(f"   ✅ Low Risk Profile - Business Performance Stable")

print(f"\n📋 STRATEGIC PRIORITIES (Next 12 Months):")
print(f"   🥇 HIGH PRIORITY:")
print(f"      • Accelerate growth in top-performing units")
print(f"      • Optimize marketing spend allocation")
print(f"      • Improve operational efficiency in underperforming units")
print(f"   🥈 MEDIUM PRIORITY:")
print(f"      • Seasonal revenue optimization strategies")
print(f"      • Customer retention and loyalty programs")
print(f"      • Technology infrastructure investments")
print(f"   🥉 ONGOING INITIATIVES:")
print(f"      • Performance monitoring and reporting systems")
print(f"      • Market expansion feasibility studies")
print(f"      • Competitive intelligence and benchmarking")

print(f"\n💰 PROJECTED FINANCIAL IMPACT:")
current_run_rate = current_ytd_revenue * (12/8)  # Annualized
optimistic_projection = current_run_rate * 1.25  # 25% improvement
conservative_projection = current_run_rate * 1.15  # 15% improvement

print(f"   • Current Annual Run Rate: ${current_run_rate:,.0f}")
print(f"   • Conservative Growth Scenario: ${conservative_projection:,.0f} (+15%)")
print(f"   • Optimistic Growth Scenario: ${optimistic_projection:,.0f} (+25%)")
print(f"   • Additional Revenue Potential: ${conservative_projection - current_run_rate:,.0f} - ${optimistic_projection - current_run_rate:,.0f}")
```
