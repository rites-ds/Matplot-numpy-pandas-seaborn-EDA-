```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(45)
```

```python
df = pd.DataFrame({
    'Age': np.random.randint(18, 60, 100),
    'Salary': np.random.randint(25000, 120000, 100),
    'Department': np.random.choice(['HR', 'Finance', 'IT', 'Marketing'], 100),
    'Experience': np.random.randint(1, 20, 100),
    'Gender': np.random.choice(['Male', 'Female'], 100)
})
```

```python
print("\nSummary Statistics:\n", df.describe())  # Numerical summary (mean, min, max, etc.)
```

```python
plt.figure(figsize=(12,5))
```

```python
# Histogram for Age
plt.subplot(1, 2, 1)  # Create subplot (row 1, col 2, position 1)
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')  # Chart title
plt.xlabel('Age')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
```

```python
# Histogram for Salary
plt.subplot(1, 2, 2)  # Position 2
plt.hist(df['Salary'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.tight_layout()  # Adjust layout to avoid overlapping
plt.show()
```

```python
plt.figure(figsize=(12, 4))

# Bar chart for Department
plt.subplot(1, 2, 1)
df['Department'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Department Count')
plt.xlabel('Department')
plt.ylabel('Count')
```

```python
# Bar chart for Gender
plt.subplot(1, 2, 2)
df['Gender'].value_counts().plot(kind='bar', color='pink', edgecolor='black')
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Age vs Salary
plt.figure(figsize=(10, 5))
plt.scatter(df['Age'], df['Salary'], color='purple', alpha=0.6)  # alpha=transparency
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Box Plot: Salary distribution per Department
plt.figure(figsize=(8, 5))
df.boxplot(column='Salary', by='Department', grid=False)
plt.title('Salary by Department')
plt.suptitle('')  # Remove default title
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
corr = df[['Age', 'Salary', 'Experience']].corr()  # Calculate correlation
plt.matshow(corr, cmap='coolwarm', fignum=1)  # Color heatmap
plt.colorbar()  # Add color scale bar
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)  # X labels
plt.yticks(range(len(corr.columns)), corr.columns)  # Y labels
plt.title('Correlation Matrix', pad=20)
plt.show()
```
